```python
#Jawaban nomer 1
class BinaryTree:
    def __init__(self,value):
        self.root = value
        self.leftChild = None
        self.rightChild  = None
        
    def getRightChild(self):
        return self.rightChild
    def getLeftChild(self):
        return self.leftChild
    def getRootVal(self):
        return self.root
    def setRootVal(self,obj):
        self.root = obj
    
    def insertLeft(self,newNode):
        if self.leftChild == None:
            self.leftChild = BinaryTree(newNode)
        else:
            t = BinaryTree(newNode)
            t.leftChild = self.leftChild
            self.leftChild = t
            
    def insertRight(self,newNode):
        if self.rightChild == None:
            self.rightChild = BinaryTree(newNode)
        else:
            t = BinaryTree(newNode)
            t.rightChild = self.rightChild
            self.rightChild = t

```

```python
a = BinaryTree('a')
a.insertLeft('b')
b = a.getLeftChild()
b.insertRight('d')
d = b.getRightChild()
a.insertRight('c')
c = a.getRightChild()
c.insertLeft('e')
e = a.getLeftChild()
c.insertRight('f')
f = c.getRightChild()

print("{:^20}".format(a.getRootVal()))
print("{:^10}{:^10}".format(b.getRootVal(),c.getRootVal()))
print("{:>8}  {:^5}{:^5}".format(d.getRootVal(),e.getRootVal(),f.getRootVal()))
```

```python
class BinaryTree:
    def __init__(self, data):
        self.data = data
        self.kiri = None
        self.kanan = None

    def evaluate(self):
        if self.data.isdigit():
            return int(self.data)
        else:
            kiri_value = self.kiri.evaluate()
            kanan_value = self.kanan.evaluate()
            if self.data == '+':
                return kiri_value + kanan_value
            elif self.data == '-':
                return kiri_value - kanan_value
            elif self.data == '*':
                return kiri_value * kanan_value
            elif self.data == '/':
                return kiri_value / kanan_value

def buildTree(postfix):
    stack = []

    # Fungsi untuk membangun pohon ekspresi dari postfix
    token = postfix.split()
    for tokens in token:
        if tokens.isdigit():
            # Jika token adalah operand, buat tree dengan node sebagai operand tersebut
            operand_tree = BinaryTree(tokens)
            stack.append(operand_tree)
        else:
            # Jika token adalah operator, pop dua tree teratas dari stack
            kanan_tree = stack.pop()
            kiri_tree = stack.pop()

            # Buat tree baru dengan root node sebagai operator
            operator_tree = BinaryTree(tokens)
            operator_tree.kanan = kanan_tree
            operator_tree.kiri = kiri_tree

            # Push tree yang terbentuk ke dalam stack
            stack.append(operator_tree)

    # Pada akhirnya, simpul akar yang tersisa di stack adalah akar dari pohon ekspresi yang dibangun
    tree = stack.pop()
    return tree

a = "2 8 9 + *"
# a = "2 4 + 3 5 * -"
# a = "10 3 2 12 + - *"
Tree = buildTree(a)

# Evaluasi pohon ekspresi
resTree = Tree.evaluate()
print("Hasil evaluasi:", resTree)
```
